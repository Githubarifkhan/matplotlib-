```python
days =list(range(0,22,3))
print(days)
```

```python
temps =[23,25,21,27,30,36,26,30]
plt.plot(days, temps)
plt.show()
```

```python
plt.plot(days, temps,'ob')
plt.show()
```

```python
temp_min =[23,25,21,27,30,36,26,30]
temp_max =[24,30,23,29,34,38,27,31]
```

```python
plt.plot(days, temp_min,'ob',
         days, temp_max,'or')
plt.show()
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.show()
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.xlabel('day')
plt.ylabel('temperature')
plt.show()
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.xticks(list(range(0,22)))
plt.show()
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(22.5,37.5)
plt.xticks(list(range(0,28,7)),['w1','w2','w3','w4'])
plt.yticks([22,25,28,31,33,37,39])
plt.show()
```

```python
legends
```

```python
plt.plot(days, temp_min,
         days, temp_min,'ob',
         days, temp_max,
         days, temp_max,'or')
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(22.5,37.5)
plt.xticks(list(range(0,28,7)),['w1','w2','w3','w4'])
plt.yticks([22,25,28,31,33,37,39])
ax.legend(['min','max'])
plt.show()
```

```python
plt.plot(days, temp_min)
plt.plot(days, temp_min,'ob',label='min')
plt.plot(days, temp_max,)
plt.plot(days, temp_max,'or',label ='max')
plt.plot(loc='upper best')
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(22.5,37.5)
plt.xticks(list(range(0,28,7)),['w1','w2','w3','w4'])
plt.yticks([22,25,28,31,33,37,39])
ax.legend(['min','max'])
plt.show()
```

```python
ANNOTATION
```

```python
plt.plot(days, temp_min)
plt.plot(days, temp_min,'ob',label='min')
plt.plot(days, temp_max,)
plt.plot(days, temp_max,'or',label ='max')
plt.plot(loc='upper best')
plt.annotate('temperature difference \n across month ',xy=(3,35))
plt.xlabel('day')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_bounds(22.5,37.5)
plt.xticks(list(range(0,28,7)),['w1','w2','w3','w4'])
plt.yticks([22,25,28,31,33,37,39])
ax.legend(['min','max'])
plt.show()
```
